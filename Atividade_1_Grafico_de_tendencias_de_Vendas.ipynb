```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
lojas_df = pd.DataFrame(vendas_2022)
lojas_df ['ID'] = lojas
lojas_df = lojas_df.set_index('ID')
lojas_df
```

```python
lojas_df.describe()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 6))
fig.suptitle('Numero de vendas em 2022', y= 1.02)
fig.subplots_adjust(hspace= 0.5)

axs[0,0].plot(lojas_df.loc['A'], color='Blue')
axs[0,0].set_title('Loja A')
axs[0,0].grid(True)

axs[0,1].plot(lojas_df.loc['B'], color='Green')
axs[0,1].set_title('Loja B')
axs[0,1].grid(True)

axs[1,0].plot(lojas_df.loc['C'], color='Red')
axs[1,0].set_title('Loja C')
axs[1,0].grid(True)

axs[1,1].plot(lojas_df.loc['D'], color='Black')
axs[1,1].set_title('Loja D')
axs[1,1].grid(True)

ymin = 0
ymax = 450
for ax in axs.ravel():
    ax.set(ylim=(ymin, ymax))

for ax in axs.flat:
    ax.set(ylabel='Numero de Vendas')

plt.show()
```
